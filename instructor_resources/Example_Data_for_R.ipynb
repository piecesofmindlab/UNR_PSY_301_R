```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
years_education = np.random.randint(low=12, high=16, size=(100,))
income = (years_education * 3000) + np.random.normal(loc=0, scale=3000, size=(100,))
plt.scatter(years_education, income)
```

```python
df = pd.DataFrame(dict(years_education=years_education, income=income))
```
